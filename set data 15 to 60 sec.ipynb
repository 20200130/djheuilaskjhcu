```python
# 먼저 모듈 불러오기
import numpy as np
import pandas as pd

# 아무 train data나 불러오기
train_data = pd.read_csv('C:/jupyter_project/train/111.csv')

# array로 행렬 만들기
arr = np.array(train_data)

# shape(행, 열) 확인
arr.shape

```

```python
# 모든 값이 동일한 열이 있는지 확인하기
# 그 전에 한 열의 모든 값 출력해보기
# 참고로 0열은 time
arr[ : ,1]
```

```python
# 이 값이 첫째 행의 값과 같은가?
arr[ : ,1] == arr[0,1]

# 모두 True라면 이 열을 삭제
if arr[ : ,1] == arr[0,1]:
    del arr[ : ,1]
    
    
    
```
